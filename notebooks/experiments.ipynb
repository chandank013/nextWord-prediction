#### Project Overview

This project focuses on building a deep learning–based **next word prediction system** using **Long Short-Term Memory (LSTM)** networks. LSTMs are particularly effective for handling sequential text data, making them well-suited for natural language processing tasks such as word prediction.

1. **Data Collection** :The dataset used for this project is the text of Shakespeare’s *Hamlet*. This rich and complex literary work provides meaningful linguistic patterns and long-term dependencies, making it an ideal dataset for training an LSTM-based language model.

2. **Data Preprocessing** :The collected text is preprocessed by tokenizing it into individual words, converting the text into numerical sequences, and applying padding to ensure uniform input lengths. The processed sequences are then divided into training and testing datasets.

3. **Model Building** : The model architecture consists of an embedding layer to represent words in dense vector form, followed by two stacked LSTM layers to capture sequential dependencies. A dense output layer with a **softmax activation function** is used to predict the probability distribution of the next word in the sequence.

4. **Model Training** :The model is trained on the prepared sequences using an appropriate loss function and optimizer. **Early stopping** is applied to monitor validation loss and halt training when performance stops improving, thereby reducing the risk of overfitting.

5. **Model Evaluation** :After training, the model is evaluated using sample input sentences to assess its accuracy in predicting the next word. This qualitative evaluation helps measure how effectively the model understands contextual relationships between words.

6. **Deployment** :To make the model interactive and user-friendly, it is deployed as a **Streamlit web application**. The application allows users to input a sequence of words and instantly receive the predicted next word in real time.
